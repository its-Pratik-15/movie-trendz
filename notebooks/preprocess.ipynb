```python
unknown_counts = (
    df
    .select_dtypes(include='object')
    .apply(lambda col: col.str.strip().str.lower().eq('unknown').sum())
    .sort_values(ascending=False)
)

unknown_counts[unknown_counts > 0]
```

```python
#converting coulmns to numeric which are stored as object/string
num_cols = [
    'Score', 'Episodes', 'Rank',
    'Scored By'
]

for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
#replacing 'UNKNOWN' or any unknown with NaN(Not a number) to handel it easily 
import numpy as np
import re

df = df.replace(
    r'(?i)^[\s\u00A0]*unknown[\s\u00A0]*$',
    np.nan,
    regex=True
)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

# Handling missing values :

```python
df.drop(columns=['Licensors', 'Producers', 'English name'], inplace=True) #too many unknown

df['Premiered_missing'] = df['Premiered'].isna().astype(int)

df['Scored By'] = df['Scored By'].fillna(0)
#no fake data for preprocessing
df['Episodes'] = df['Episodes'].fillna(df['Episodes'].median()) 

df['Score_missing'] = df['Score'].isna().astype(int)

df.dropna(subset=['Rank', 'Score'],inplace=True)

df['Rating'] = df['Rating'].fillna(df['Rating'].mode()[0])

df['Type'] = df['Type'].fillna(df['Type'].mode()[0])

df['Other name'] = df['Other name'].fillna('Not Available')

df['Studios'] = df['Studios'].fillna('Unknown_Studio')

df['Genres'] = df['Genres'].fillna('Unknown_Genre')
```

```python
df.isnull().sum().sort_values(ascending=False)
```

# Anime Score Distribution

```python
plt.hist(df['Score'].dropna(), bins=20)
plt.title("Anime Score Distribution")
plt.xlabel("Score")
plt.ylabel("Count")
plt.show()
```

```python
df['Rating'].value_counts().plot(kind='bar')
plt.title("Age Rating Distribution")
plt.show()
```

```python
genres = df['Genres'].str.split(',').explode().str.strip()
genres.value_counts().head(15).plot(kind='bar')
plt.title("Top 15 Anime Genres")
plt.show()
```

```python
plt.hist(df['Members'], bins=30)
plt.title("Members Distribution")
plt.show()
```

```python
plt.scatter(df['Popularity'], df['Score'], alpha=0.4)
plt.xlabel("Popularity Rank")
plt.ylabel("Score")
plt.title("Score vs Popularity")
plt.show()
```

```python
num_cols = [
    'Score', 'Episodes', 'Rank',
    'Popularity', 'Favorites',
    'Scored By', 'Members'
]
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
df['Year'] = df['Aired'].str.extract(r'(\d{4})').astype(float)
df['Year'].value_counts().sort_index().plot()
plt.title("Anime Releases Over Time")
plt.show()
```
